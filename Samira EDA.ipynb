# Breakdown Causes/Payment Type

I think that payment type will be beneficial for our analysis because it allows us to look at whether a water well is invested in rather than simply paid for at one fixed time. The reasoning behind whether it is paid for throughout a long period of time allows us to see if it is functioning better because there is constant maintenance. This also goes hand in hand with funder vs. non funder.


(fromhttps://snv.org/update/functionality-rural-water-sources-largely-dependent-functional-professional-systems) A reliable water source should provide water for a minimum of 350 days in a year with less than 14 days of breakdown. Very few water sources are able to meet this standard and a broken down water source can take up to three months or more before it is repaired. On average, a water source functions well within the first three years, after which it starts breaking down. The majority of the current rural water sources are over 5 years old, not regularly maintained and hence at risk of regular breakdowns.
